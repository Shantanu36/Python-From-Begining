```python
import numpy as np
import matplotlib.pyplot as plt
from io import StringIO,BytesIO
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel("x axis")
plt.ylabel('y axis')
plt.title('Graph in 2D')
```

```python
plt.plot(x,y,'r')
plt.xlabel("x axis")
plt.ylabel('y axis')
plt.title('Plot in 2D')
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'r*-')
```

```python
plt.plot(x,y,'ro',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('2-D Graph')
```

### Creating Subplot

### subplots of sine and cosine wave

```python
x=np.arange(0,5* np.pi,0.2)
y_sin=np.sin(x)
y_cos=np.cos(x)
```

```python
#1st subplot
plt.subplot(2,2,1)
plt.plot(x,y_sin,c='g')
plt.title('sine wave')

#2nd subplot
plt.subplot(2,2,2)
plt.plot(x,y_cos)
plt.title('cos wave')
```

### BOX Plot

```python
data=[np.random.normal(0,std,100) for std in range(1,5)]
```

## Pie chart

```python
labels='Pyhon','R','JAVA','SQL','DW'
sizes=[100,50,70,150,175]
colors=['red','green','blue','orange','pink']
explode=(0.1,0,0,0,0) #explodes 1st size
```

```python
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
```
